```python
# "%matplotlib inline"  renders the figure in a notebook 
%matplotlib inline                    
import matplotlib.pyplot as plt
#plt.style.use('seaborn-whitegrid')    # defines a form of style sheet (see https://matplotlib.org/ for list of styles)
plt.style.use('dark_background')
import numpy as np

# lets use "linspace" function which returns evenly spaced numbers over the specified interval
array = np.linspace(0, 10, num=100)       # refers np.linspace(start=0, stop=10, num=100)
plt.plot(array, np.sin(array));           # define a sine wave for array variable

plt.plot(array, np.cos(array),color='blue', linestyle='solid', linewidth='3')   # plot cosine with a blue continuous line of width 3
plt.plot(array, np.sin(array + 30), color='g', linestyle='--')
plt.plot(array, np.sin(array + 45), color='c', linestyle='-.')
plt.plot(array, np.sin(array + 60), color='y', linestyle='dotted')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Make some fake data.
y = np.arange(2, 50, 2)

# Create plots with pre-defined labels.
fig, ax = plt.subplots()
ax.plot(y, np.log(y*y), 'r*', label='y*y')
ax.plot(y, np.log(y*y**2), 'b^', label='y*y**2')
ax.plot(y, np.log(y*y**3), 'gs', label='y*y**3')

legend = ax.legend(loc='upper left', shadow=True, fontsize='small')  # best

# Put a nicer background color on the legend.
legend.get_frame().set_facecolor('C5')

plt.show()
```

### _**Subplot** in matplotlib_

A "figure" represents the whole window in the user interface, meanwhile, there can be **"subplots"** in different sizes. These **subplots** can be arranged in regular grid with the specified number of columns and rows. 

![](./Figures/subplot.png)


```python
# lets draw a signle plot by subplot
%matplotlib inline

x = np.linspace(0, 3 * np.pi, 300)
y = np.sin(x**2)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('A single plot')
plt.show()
```

```python
# Lets subplot a time series of random numbers

fig, (ax1, ax2) = plt.subplots(2, 1)
fig.subplots_adjust(hspace=0.5)       # add an extra space between the subplots

t = np.arange(0, 30, 0.01)
s1 = 0.01 * np.sin(2 * np.pi*t)   # white noise added
s2 = 0.01 * np.sin(2 * np.pi*t) + np.random.randn(len(t))   # white noise added

ax1.plot(t, s1, c='goldenrod')

ax1.set_xlim(0, 5)
ax1.set_xlabel('time')
ax1.set_ylabel('sin')
#ax1.set_ylim(-3,3 )

ax2.set_xlim(0, 5)
ax2.set_xlabel('time')
ax2.set_ylabel('noise')

ax2.plot(t, s2, c='darkslategray')
plt.show()
```

```python
# use sharing axes 
fig, axs = plt.subplots(2, sharex=True, sharey=False, gridspec_kw={'hspace': 0})
fig.suptitle('Sharing axes')

axs[0].plot(t, s1)
axs[1].plot(t, s2)

plt.show()
```

#### Customize more complicated arrangements can also be perform by using "GridSpec", "subplotspec" or "subplot2grid" modules


```python
# Customize more complicated arrangements can also be perform by using "GridSpec", "subplotspec" or "subplot2grid" module
import matplotlib.gridspec as gridspec

grid = plt.GridSpec(ncols=3, nrows=3, wspace=0.5, hspace=0.3)

plt.subplot(grid[0, 0])
plt.subplot(grid[0, 1:])
plt.subplot(grid[1, :2])
plt.subplot(grid[1, 2])
plt.subplot(grid[-1, -2])
plt.show()
```
